```python
from rdkit.Chem import Descriptors
```

```python
fces = (Descriptors.MolWt, Descriptors.NumHeteroatoms, Descriptors.RingCount)
```

```python
ds = [tuple(f(mol) for f in fces) for mol in mols]
```

```python
sums = [0,0,0]
for mw, nh, nr in ds:
    sums[0] += mw
    sums[1] += nh
    sums[2] += nr
        
```

```python
[s/len(mols) for s in sums]
```
